```python
active_times = []
for idx in sources.index:
    name = (sources.iloc[idx]['name'])
    name = name.replace(' ', '')
    name = name.replace('+', 'p')
    active_times.append(time_active[name])

```

```python
base_dir = '/data/user/ssclafani/data/analyses/XRB_baseline_v1.0/combo/lc/'
```

```python
sens_dict = {}
ra_dict = {}
dec_dict = {}

for name in sorted(sources.name_disp):
    ra = sources[sources.name_disp == name].ra_deg.values[0]
    dec = sources[sources.name_disp == name].dec_deg.values[0]
    #print(ra,dec)
    try:
        sens = np.load(base_dir + f'{name}/trials/fit_gamma/no_gpinj/src_gamma_2.0/thresh_0.0/lag_0.0/sens.npy', 
                   allow_pickle = True)[()]
        #print(sens)
        sens_dict[name] = sens['flux_E2dNdE_1TeV']
        ra_dict[name] = ra
        dec_dict[name] = dec
    except:
        print(f'{name} not found')

sens_dict_E3 = {}
ra_dict_E3 = {}
dec_dict_E3 = {}
for name in sorted(sources.name_disp):
    ra = sources[sources.name_disp == name].ra_deg.values[0]
    dec = sources[sources.name_disp == name].dec_deg.values[0]
    #print(ra,dec)
    try:
        sens = np.load(base_dir + f'{name}/trials/fit_gamma/no_gpinj/src_gamma_3.0/thresh_0.0/lag_0.0/sens.npy', 
                   allow_pickle = True)[()]
        #print(sens)
        sens_dict_E3[name] = sens['flux_E2dNdE_1TeV']
        ra_dict_E3[name] = ra
        dec_dict_E3[name] = dec
    except:
        print(f'{name} not found')
```

```python
dp_dict = {}
radp_dict = {}
decdp_dict = {}

for name in sorted(sources.name_disp):
    ra = sources[sources.name_disp == name].ra_deg.values[0]
    dec = sources[sources.name_disp == name].dec_deg.values[0]
    #print(ra,dec)
    try:
        sens = np.load(base_dir + f'{name}/trials/fit_gamma/no_gpinj/src_gamma_2.0/thresh_0.0/lag_0.0/5.0sigma_dp.npy', 
                   allow_pickle = True)[()]
        #print(sens)
        dp_dict[name] = sens['flux_E2dNdE_1TeV']
        radp_dict[name] = ra
        decdp_dict[name] = dec
    except:
        print(f'{name} not found')

dp_dict_E3 = {}
radp_dict_E3 = {}
decdp_dict_E3 = {}
for name in sorted(sources.name_disp):
    ra = sources[sources.name_disp == name].ra_deg.values[0]
    dec = sources[sources.name_disp == name].dec_deg.values[0]
    #print(ra,dec)
    try:
        sens = np.load(base_dir + f'{name}/trials/fit_gamma/no_gpinj/src_gamma_3.0/thresh_0.0/lag_0.0/5.0sigma_dp.npy', 
                   allow_pickle = True)[()]
        #print(sens)
        dp_dict_E3[name] = sens['flux_E2dNdE_1TeV']
        radp_dict_E3[name] = ra
        decdp_dict_E3[name] = dec
    except:
        print(f'{name} not found')
```

```python
prev_sens = np.genfromtxt('/home/ssclafani/XRB_Analysis/icecube_table3.csv', delimiter=',', skip_header=1)
```

```python
sens_decs = np.array([float(d) for d in dec_dict.values()])
sens_flux = np.array([float(f) for f in sens_dict.values()])

sens_decs_E3 = np.array([float(d) for d in dec_dict_E3.values()])
sens_flux_E3 = np.array([float(f) for f in sens_dict_E3.values()])


dp_decs_E3 = np.array([float(d) for d in decdp_dict_E3.values()])
dp_flux_E3 = np.array([float(f) for f in dp_dict_E3.values()])

dp_decs = np.array([float(d) for d in decdp_dict.values()])
dp_flux = np.array([float(f) for f in dp_dict.values()])
```

```python
ra = prev_sens[:,0]
dec = prev_sens[:,1]
flux = prev_sens[:,-3]
flux_E3 = prev_sens[:,-1]


ratio = {}
for i, d in enumerate(dec):
    mask = np.isclose(sens_decs, d, atol=.1)
    if sum(mask) == 1:
        ratio[d] =  sens_flux[mask][0]/(flux[i]*1e-3)

ratio_E3 = {}
for i, d in enumerate(dec):
    mask = np.isclose(sens_decs_E3, d, atol=.1)
    if sum(mask) == 1:
        ratio_E3[d] =  sens_flux_E3[mask][0]/(flux_E3[i]*1e-3)

dp_ratio = {}
for i, d in enumerate(dec):
    mask = np.isclose(dp_decs, d, atol=.1)
    if sum(mask) == 1:
        dp_ratio[d] =  dp_flux[mask][0]/(flux[i]*1e-3)
        
dp_ratio_E3 = {}
for i, d in enumerate(dec):
    mask = np.isclose(dp_decs_E3, d, atol=.1)
    if sum(mask) == 1:
        dp_ratio_E3[d] =  dp_flux_E3[mask][0]/(flux_E3[i]*1e-3)
        
```

```python
norm = plt.Normalize((min(np.log(active_times))), (max(np.log(active_times))))
colors = plt.cm.viridis(norm(np.log(active_times)))  # Use the 'viridis' colormap
```
